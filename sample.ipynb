```python
#import required package

import pandas as pd
import numpy as np
```

```python

df=pd.read_csv("all_matches.csv")
df.head()
```

```python
#Find missing values
df.isnull().sum()
```

```python
#fill missing values with 0
df.fillna(0,inplace=True)
```

```python
#Check is there any missing values
df.isnull().sum()
```

```python
df.head()
```

```python
#check columns
df.columns
```

```python
#create new column
df['total_run']=df['runs_off_bat']+df['extras']
```

```python
#Runs scores by Batsman
df.groupby('striker').runs_off_bat.sum().sort_values(ascending=False)
```

```python
#Check unique values
df['season'].unique()
```

```python
#in season we see some duplicate values, we are going to replace it
df['season'].replace([2012,2013,2014,2015,2016,2017,2018,2019,2021],['2012','2013','2014','2015','2016','2017','2018','2019','2021'],inplace=True)
```

```python
#Runs scored by team by season
ssn=df.groupby(['season','batting_team']).total_run.sum().reset_index()
ssn.head(8)
```

```python
season_1=ssn[ssn['season']=='2007/08'].sort_values(by='total_run',ascending=False)
season_1
```

```python
#plot a chart
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=season_1['batting_team']
y=season_1['total_run']

plt.bar(x,y)
plt.show()
```
